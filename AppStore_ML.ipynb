```python
import pandas as pd
import numpy as np
import matplotlib as plt
import sklearn as sk
dataset = pd.read_csv('./AppleStore.csv')
```

```python
dataset.head()
```

```python
num_c = [
    'size_bytes',
    'price',
    'rating_count_tot',
    'rating_count_ver',
    'sup_devices.num',
    'ipadSc_urls.num',
    'lang.num',
    'cont_rating',
]

cat_c = [
    'currency',
    'prime_genre'
]

target_c = ['user_rating']

cols = num_c + cat_c + target_c
data = dataset[cols]
```

```python

data.head()
```

```python
data.shape
```

```python
data.drop_duplicates()
```

```python
data['cont_rating'] = data['cont_rating'].str.slice(0,-1).apply(int)
```

```python
data.head()
```

```python
data.isna().mean()
```

```python
for i in cat_c:
  print(f"{i}")
  print(data[i].value_counts())
  print()
```

```python
data = data.drop(columns = ['currency'])
cat_c.remove('currency')
cols.remove('currency')
```

```python
data.corr()
```

```python
target_c
```

```python
for i in num_c:
  print(f"{i}", end='\t')
  print(data[i].mean(), data[i].max(), data[i].min())
```

```python
data['is_free'] = data['price'] == 0
cat_c.append('is_free')
cols.append('is_free')
```

```python
data.corr()
```

```python
data = pd.get_dummies(data, columns=cat_c) 
```

```python
data.head()
```

```python
cat_c_new = []
for i in cat_c:
  cat_c_new.extend(filter(lambda x: x.startswith(i), data.columns))
cat_c = cat_c_new
cols.extend(cat_c_new)
```

```python
cols
```

```python
data.corr()
```

```python
from sklearn.preprocessing import StandardScaler
pca = StandardScaler()
```

```python
X = pca.fit_transform(data[num_c + cat_c])
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_t, y_tr, y_t = train_test_split(X, data[target_c], test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
def score(y_pred, y):
  print('Test: '+str(i))
  print(f"R^2: {r2_score(y, y_pred)}")
  print(f"MSE: {mean_squared_error(y, y_pred)}")
  print()
  
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_tr, y_tr)
score(lr.predict(X_t), y_t)
```

```python
knn = KNeighborsRegressor(n_neighbors=1)
```

```python

knn = KNeighborsRegressor(n_neighbors=30)
knn.fit(X_tr, y_tr)
score(knn.predict(X_t), y_t)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, make_scorer
from sklearn.model_selection import cross_validate
```

```python
score_arr = cross_validate(LinearRegression(), X, data[target_c], cv=5, 
               scoring={'r2_score': make_scorer(r2_score), 
                        'mean_squared_error': make_scorer(mean_squared_error)})
```

```python
print('R^2: '+ str(score_arr['test_r2_score'].mean()))
print('MSE: '+ str(score_arr['test_mean_squared_error'].mean()))
```
