```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./forest_dataset.csv')
data.head()
```

```python
results = data[data.columns[-1]].values
features = data[data.columns[:-1]].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, results, test_size = 0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
grid = {
    'n_neighbors': np.arange(1,11),
    'metric': ['manhattan', 'euclidean'],
    'weights': ['uniform', 'distance'],
}
search = GridSearchCV(knn, grid, cv=5, scoring = 'accuracy', n_jobs = -1)
search.fit(features, results)
search.best_params_
```

```python
knn = KNeighborsClassifier(n_neighbors = 4, metric = 'manhattan', weights = 'distance')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
pred_prob = knn.predict_proba(X_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

unique, freq = np.unique(y_test, return_counts=True)
freq = list(map(lambda x: x / len(y_test),freq))

pred_freq = pred_prob.mean(axis=0)
plt.figure(figsize=(10, 8))
plt.bar(range(1, 8), pred_freq, width=0.4, align="edge", label='prediction')
plt.bar(range(1, 8), freq, width=-0.4, align="edge", label='real')
plt.ylim(0, 0.54)
plt.legend()
plt.show()
```
